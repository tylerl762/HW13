```python
import sqlite3
import pandas as pd
connection = sqlite3.connect('books.db')
```

```python
pd.read_sql("SELECT last FROM authors ORDER BY last DESC", connection)
```

```python
pd.read_sql("SELECT title FROM titles ORDER BY title ASC", connection)
```

```python
pd.read_sql("""SELECT title, first, last 
            FROM titles 
            INNER JOIN author_isbn 
                ON titles.isbn = author_ISBN.isbn
            INNER JOIN authors 
                ON author_isbn.id = authors.id
            WHERE first LIKE 'Abbey'
            """, connection)
```
